### ***MV portfolio***

```python

# 1) Numeric returns matrix (DataFrame) -> μ and Σ as NumPy
R_df = df.select_dtypes('number').drop(columns=['QAI'], errors='ignore')
mu  = R_df.mean().to_numpy(dtype=float).reshape(-1, 1)      # (N,1)
Sigma = R_df.cov().to_numpy(dtype=float)                    # (N,N)

# 2) Invert covariance (robust)
try:
    S = np.linalg.inv(Sigma)
except np.linalg.LinAlgError:
    S = np.linalg.pinv(Sigma)

# 3) Tangency weights (rf=0), normalized to 1' w = 1
ones_row = np.ones((1, mu.shape[0]))                        # (1,N)
w_num = S @ mu                                              # Σ^{-1} μ
w_den = float(ones_row @ w_num)                             # 1' Σ^{-1} μ
w_t = (w_num / w_den).reshape(-1, 1)                        # (N,1)

w_t_series = pd.Series(w_t.ravel(), index=R_df.columns, name='Tangency')
print(w_t_series.round(4).sort_values(ascending=False).to_string())
print("Sum of weights:", round(w_t_series.sum(), 6))

# 4) Portfolio series + Sharpe
p_tan = pd.Series((R_df.to_numpy() @ w_t).ravel(), index=R_df.index, name='Tangency')
m, v = p_tan.mean(), p_tan.std(ddof=0)
print(f"Tangency portfolio (monthly): mean={m:.4f}, vol={v:.4f}, Sharpe={m/v:.3f}")

# 5) Per-asset Sharpe (for quick comparison)
stats = pd.DataFrame({'Mean': R_df.mean(), 'Vol': R_df.std(ddof=0)})
stats['Sharpe'] = stats['Mean'] / stats['Vol']
print("\nAsset Sharpe ratios (desc):")
print(stats['Sharpe'].round(3).sort_values(ascending=False).to_string())

#
print("Ranking of weights do NOT align with the ranking of Sharpe ratios")
```

## 4. TIPS

**Task:** Assess how the tangency portfolio (and its performance) change under two scenarios.

### Scenarios
1. **Drop TIPS** completely from the investment set.  
2. **Bump TIPS expected excess return** by **+0.0012** relative to its historical sample mean.

### Report
- **Weights:** Baseline vs. *No TIPS* vs. *TIPS μ + 0.0012*
- **Performance (monthly):** Mean, Volatility, Sharpe

### Discussion
- How much do weights change vs. baseline (e.g., L1 norm across common assets)?
- Does the TIPS weight rise when μ is bumped?
- Do TIPS expand the investment opportunity set, i.e., should Harvard treat TIPS as a separate asset?



```python
R = df.select_dtypes('number').drop(columns=['QAI'], errors='ignore')
TIPS_COL = 'TIP'

def tangency(R, mu_override=None):
    mu = R.mean() if mu_override is None else mu_override.reindex(R.columns)
    S  = np.linalg.pinv(R.cov().to_numpy())
    w  = S @ mu.values
    w  = w / (np.ones(len(w)) @ w)                      # normalize 1'w=1
    p  = R.to_numpy() @ w
    return (pd.Series(w, index=R.columns),
            pd.Series({'Mean': p.mean(), 'Vol': p.std(ddof=0), 'Sharpe': p.mean()/p.std(ddof=0)}))

# Baseline
w0, perf0 = tangency(R)

# Drop TIPS
if TIPS_COL in R.columns:
    w_drop, perf_drop = tangency(R.drop(columns=[TIPS_COL]))
    l1_change = (w0.drop(TIPS_COL) - w_drop).abs().sum()
else:
    w_drop, perf_drop, l1_change = None, None, np.nan

# Bump TIPS expected excess return by +0.0012
if TIPS_COL in R.columns:
    mu_adj = R.mean()
    mu_adj.loc[TIPS_COL] += 0.0012
    w_bump, perf_bump = tangency(R, mu_override=mu_adj)
else:
    w_bump, perf_bump = None, None

# Report
summary = pd.concat({
    'Baseline': perf0.round(4),
    'No TIPS':  (perf_drop.round(4) if perf_drop is not None else pd.Series({'Mean':np.nan,'Vol':np.nan,'Sharpe':np.nan})),
    'TIPS μ+0.0012': (perf_bump.round(4) if perf_bump is not None else pd.Series({'Mean':np.nan,'Vol':np.nan,'Sharpe':np.nan}))
}, axis=1)
print(summary.to_string())

if TIPS_COL in R.columns:
    print(f"\nTIPS weight — baseline: {w0[TIPS_COL]:.3f} | with μ+0.0012: {w_bump[TIPS_COL]:.3f}")
    print(f"L1 change across common assets when dropping TIPS: {l1_change:.3f}")

```

### ***Equally-weighted (EW)***

Rescale the entire weighting vector to have the **target mean**. Thus, the element of the weight vector is

$$
w_i \;=\; \frac{1}{N}, 
\qquad i=1,\ldots,N
$$


- $N$: number of assets  


```python
# Equal-weighted portfolio from numeric asset columns
asset_cols = df.select_dtypes(include="number").columns  # choose only numeric assets
# If you need to drop a column (e.g., a benchmark), do: asset_cols = asset_cols.drop(["QAI"], errors="ignore")

w_eq = 1.0 / len(asset_cols)          # equal weight per asset
eq_port = (df[asset_cols] * w_eq).sum(axis=1)   # portfolio time series (as a pandas Series)
    
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(11, 6))

# Plot all assets faintly
df[asset_cols].plot(ax=ax, linewidth=1, alpha=0.35, legend=False)

# Overlay highlighted EQ portfolio
eq_line = eq_port.plot(ax=ax, linewidth=3, color="black", label="Equal-Weighted (EQ_Port)", zorder=10)

ax.set_title("All Assets + Highlighted Equal-Weighted Portfolio")
ax.set_xlabel("Date")
ax.set_ylabel("Excess Return")
ax.grid(True, alpha=0.3)
ax.legend()
plt.tight_layout()
plt.show()
```

### ***Risk-parity***

Risk-parity is a term used in a variety of ways, but here we have in mind setting the weight of the portfolio to be proportional to the inverse of its full-sample variance estimate. Thus, the element of the weight vector is,

$$
w^{RP}_i \;=\; \frac{1}{\sigma_t^2}, 
\qquad i=1,\ldots,N
$$




```python
# --- Risk-parity (inverse-variance) portfolio — ignore look-ahead bias ---
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patheffects as pe

# Full-sample inverse-variance weights
var = df[asset_cols].var(skipna=True)          # sample variance over entire sample
inv_var = 1.0 / var.replace(0, np.nan)
w_rp = (inv_var / inv_var.sum()).fillna(0)     # normalize to sum to 1
print("Risk-parity weights:\n", w_rp.sort_values(ascending=False).round(4))

# RP portfolio time series
rp_port = df[asset_cols].mul(w_rp, axis=1).sum(axis=1)

# Plot: all assets faint, RP portfolio highlighted
fig, ax = plt.subplots(figsize=(11, 6))
df[asset_cols].plot(ax=ax, linewidth=1, alpha=0.35, legend=False)
line, = ax.plot(rp_port.index, rp_port.values, linewidth=3, color="black", label="Risk-Parity (RP)", zorder=10)
line.set_path_effects([pe.Stroke(linewidth=5, foreground="white"), pe.Normal()])

ax.set_title("All Assets + Highlighted Risk-Parity Portfolio (full-sample inverse-variance)")
ax.set_xlabel("Date")
ax.set_ylabel("Return")
ax.grid(True, alpha=0.3)
ax.legend()
plt.tight_layout()
plt.show()
```

All portfolios have different expected returns, so rescaling the returns into 0.01 makes us able to compare only volatility

```python
# Make mixed-format weights (ndarray, scalar, Series) work and compare at μ* = 0.01

import numpy as np, pandas as pd
mu_star = 0.01
R = df.select_dtypes('number').drop(columns=['QAI'], errors='ignore')
mu = R.mean()

def as_series(w, cols):
    if isinstance(w, pd.Series):
        return w.reindex(cols).astype(float)
    arr = np.asarray(w, dtype=float)
    if arr.ndim: arr = arr.reshape(-1)
    if arr.size == 1:    # scalar -> equal weight per asset
        return pd.Series(float(arr.item()), index=cols)
    if arr.size != len(cols):
        raise ValueError("weights length mismatch")
    return pd.Series(arr, index=cols)

def scale_to_target(w_ser):
    k = mu_star / float(mu @ w_ser)
    return w_ser * k

w_mv_s = scale_to_target(as_series(w_t, R.columns))
w_eq_s = scale_to_target(as_series(w_eq, R.columns))
w_rp_s = scale_to_target(as_series(w_rp, R.columns))

port_mv = R @ w_mv_s
port_eq = R @ w_eq_s
port_rp = R @ w_rp_s

def perf(x): m=x.mean(); v=x.std(ddof=0); return pd.Series({'Mean':m,'Vol':v,'Sharpe':m/v})
summary = pd.DataFrame({
    'MV (μ*=0.01)': perf(port_mv),
    'EW (μ*=0.01)': perf(port_eq),
    'RP (μ*=0.01)': perf(port_rp)
}).T.round(4)

print(summary.to_string())

```
