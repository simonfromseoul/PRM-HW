```python
# load data

import pandas as pd
import numpy as np
from pathlib import Path

file_path = Path("proshares_analysis_data.xlsx")
df_hf_etf = pd.read_excel(file_path, engine="openpyxl", sheet_name="hedge_fund_series")

# Peek at the data
print(df_hf_etf.shape)
df_hf_etf.head()
etf_cols = df_hf_etf.select_dtypes(include="float").columns.tolist()
print(etf_cols)
```

## 1. Find mean, vol and Sharpe

```python
rets = df_hf_etf[etf_cols]
k = 12  # monthly → annual

mean_m = rets.mean()
vol_m  = rets.std(ddof=1)

print("-- Mean (annualized) --")
print(mean_m * k)

print("\n-- Vol (annualized) --")
print(vol_m * np.sqrt(k))

print("\n-- Sharpe (annualized) --")
print((mean_m / vol_m) * np.sqrt(k))
```

## 2. 

For the series in the “hedge fund series” tab, calculate the following statistics related to tail-risk.

- **Skewness**

- **Excess Kurtosis** (in excess of 3)

- **VaR (.05)** - the fifth quantile of historic returns

- **CVaR (.05)** - the mean of the returns at or below the fifth quantile

- **Maximum drawdown** - include the dates of the max/min/recovery within the max drawdown period.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# If your datetime is a column (possibly unnamed), set it as index once:
if not isinstance(df_hf_etf.index, pd.DatetimeIndex):
    dt_col = df_hf_etf.select_dtypes(include=["datetime"]).columns[0]
    df_hf_etf = df_hf_etf.set_index(dt_col).sort_index()

rows = []
for col in etf_cols:
    s = df_hf_etf[col]

    # Tail-risk stats
    skew = s.skew()
    ex_kurt = s.kurt()                  # excess kurtosis (0 for normal)
    var05 = s.quantile(0.05)            # 5th percentile
    cvar05 = s[s <= var05].mean()       # CVaR at 5%


    rows.append([col, skew, ex_kurt, var05, cvar05])


tail_stats = pd.DataFrame(
    rows,
    columns=[
        "Series", "Skewness", "Excess Kurtosis",
        "VaR (0.05)", "CVaR (0.05)"]
).set_index("Series")

print(tail_stats)
```

```python
# Maximum Draw Down

rows = []
for col in etf_cols:
    s = df_hf_etf[col]
    wealth = (1 + s).cumprod()
    peak_run = wealth.cummax()
    ratio = wealth / peak_run

    trough_date = ratio.idxmin()
    peak_date = wealth.loc[:trough_date].idxmax()
    peak_val = wealth.loc[peak_date]
    mdd = ratio.loc[trough_date] - 1

    rec = wealth.loc[trough_date:]
    recovery_date = rec.index[(rec >= peak_val).argmax()] if (rec >= peak_val).any() else pd.NaT

    rows.append([col, mdd, peak_date, trough_date, recovery_date])

    plt.figure()
    plt.plot(wealth.index, wealth.values, label=col)
    plt.scatter([peak_date], [wealth.loc[peak_date]])
    plt.scatter([trough_date], [wealth.loc[trough_date]])
    if not pd.isna(recovery_date):
        plt.scatter([recovery_date], [wealth.loc[recovery_date]])
    plt.title(f"{col} – MDD")
    plt.xlabel("Date"); plt.ylabel("Cumulative wealth")
    plt.legend(); plt.grid(True); plt.show()

mdd_stats = pd.DataFrame(rows, columns=["Series","Max Drawdown","Peak Date","Trough Date","Recovery Date"]).set_index("Series")
print(mdd_stats)

```

# 3.
For the series in the “hedge fund series” tab, run a regression of each against **SPY** (found in the “merrill factors” tab).  
Include an intercept. Report the following regression-based statistics:

- **Market Beta**

- **Treynor Ratio**

- **Information Ratio**

Annualize these three statistics as appropriate.


```python
from sklearn.linear_model import LinearRegression

# Assume excess return =0.01
mu_tilda = 0.01

# Ensure both frames have a DatetimeIndex
if not isinstance(df_hf_etf.index, pd.DatetimeIndex):
    dt_col_hf = df_hf_etf.select_dtypes(include=["datetime"]).columns[0]
    df_hf_etf = df_hf_etf.set_index(dt_col_hf).sort_index()

merril_factors = pd.read_excel(file_path, engine="openpyxl", sheet_name="merrill_factors")

if not isinstance(merril_factors.index, pd.DatetimeIndex):
    dt_col_mf = merril_factors.select_dtypes(include=["datetime"]).columns[0]
    merril_factors = merril_factors.set_index(dt_col_mf).sort_index()

spy = merril_factors['SPY US Equity']  # make sure this is the returns series (not prices)

linear_model = LinearRegression(fit_intercept=True)

for col in etf_cols:
    s = df_hf_etf[col]

    # align by date and drop any missing
    pair = pd.concat([spy.rename("SPY"), s.rename("y")], axis=1, join="inner").dropna()

    X = pair[["SPY"]].to_numpy()  # shape (n,1)
    y = pair["y"].to_numpy()

    linear_model.fit(X, y)
    beta = float(linear_model.coef_[0])
    alpha = float(linear_model.intercept_)

    pred   = linear_model.predict(X)
    resid  = y - pred
    std_resid = resid.std(ddof=1) 
    
    rf = 0.01
    treynor = 12*(y.mean() - rf) / beta 
    ir = np.sqrt(12) * alpha / std_resid
    print(f"{col}: beta={beta:.6f}, alpha={alpha:.6f}, Treynor (A)={treynor:.6f}, Information Ratio (A)= {ir:.6f}, n={len(y)}")
    

```

# 7. Out of Sample test

```python
window = 60
y = df_hf_etf['HFRIFWI Index']
X = df_merril_factors.drop(columns=df_merril_factors.columns[0])  # drop datetime col
X = sm.add_constant(X)

oos_preds = []

for t in range(window, len(y)):
    y_train = y.iloc[t-window:t]
    X_train = X.iloc[t-window:t]
    model = sm.OLS(y_train, X_train).fit()
    
    X_t = X.iloc[t]
    y_hat_t = model.predict(X_t)
    oos_preds.append((y.index[t], y_hat_t))

df_oos = pd.DataFrame(oos_preds, columns=['Date', 'OOS_Replication']).set_index('Date')
df_oos['Target'] = y.loc[df_oos.index]

# Compare performance
corr = df_oos['OOS_Replication'].corr(df_oos['Target'])
print("OOS Replication Correlation:", corr)
```

```python

oos_aligned = pd.concat([
    pd.Series([np.nan]*60, index=df_hf_etf.index[:60]),
    pd.Series(df_oos['OOS_Replication'].values, index=df_hf_etf.index[60:])
])

plt.figure(figsize=(10,5))
plt.plot(df_hf_etf.index, df_hf_etf['HFRIFWI Index'], label='HFRI (Actual)')
plt.plot(oos_aligned.index, oos_aligned, label='OOS Replication', linestyle='--')
plt.legend()
plt.title("HFRI vs Out-of-Sample Replication")
plt.show()
```

### Answer for 7:
Given the correlation of 0.9 between the OOS predictions and the real values, it can be concluded that the OOS replication performs well.
